**Cancer Prediction**

```python
import pandas as pd
```

```python
cancer = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Cancer.csv')
```

```python
cancer.head()
```

```python
cancer.info()
```

```python
cancer.describe()
```

```python
cancer.columns
```

```python
y= cancer['diagnosis']
```

```python
x= cancer.drop(['id','diagnosis','Unnamed: 32'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y, train_size= 0.7, random_state= 2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
x_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model= LogisticRegression(max_iter=5000)
```

```python
model.fit(x_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

data visualization

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Assume 'data' is your dataset with relevant cancer prediction features
# You can replace it with your actual dataset
# For simplicity, let's assume you have features 'age' and 'prediction_score'
data = {'age': [30, 40, 50, 60, 70],
        'prediction_score': [0.2, 0.5, 0.7, 0.3, 0.9]}

# Create a DataFrame (you can replace this with your actual DataFrame)
import pandas as pd
df = pd.DataFrame(data)

# Scatter plot for age and prediction score
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='prediction_score', data=df)
plt.title('Cancer Prediction Visualization')
plt.xlabel('Age')
plt.ylabel('Prediction Score')
plt.show()
```

```python
y_predict= model.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import accuracy_score ,confusion_matrix, classification_report
```

```python
confusion_matrix(y_test, y_predict)
```

```python
print(classification_report(y_test,y_predict))
```
